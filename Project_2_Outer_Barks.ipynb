First, we will begin creating a random two dimensional walk by creating 2 lists for the changes in the x direction and changes in the y direction, and then making a loop that runs the random choice of directions for a walk and plot these two lists as a walk.

```python
#1
import numpy as np
import matplotlib.pyplot as plt
def proj(n):
  #n is number of steps
  k=0
  a=0
  #to start our x,y coords
  #then make a list that will contain our x,y coords, and start with 0 already inside to signify starting at the origin
  K=[0]
  A=[0]
  D=[] #this is for storing our choices
  for i in range(n):
    choice=np.random.choice(['U','D','L','R']) #this gives a random choice between up down left or right
    if choice=='U':
      a+=1 #when the random choice is 'U' we add one to our Y axis to simulate the person taking a step forward
    elif choice=='D':
      a-=1 #when the random choice is 'D' we subtract one from our Y axis to simulate the person taking a step backward
    elif choice=='L':
      k-=1 #when the random choice is 'L' we subtract one from our X axis to simulate the person is taking a step left
    elif choice=='R':
      k+=1 #when the random choice is 'R' we add one to our X axis to simulate the person is taking a step right
    K.append(k)
    A.append(a) 
    D.append(choice) #after each loop we then add the changes to our lists of x and y values as well as a list containing the direction of each step
  print(K)
  print(A)
  print(D) #these print statements are to make sure everything is working correctly
  plt.xlabel('Steps in X direction')
  plt.ylabel('Steps in Y direction') #our x,y labels
  plt.plot(K,A,c='blue',label='walking path') #plotting our walking path
  plt.scatter(0,0,alpha=1,c='orange',label='start',marker='x') 
  plt.scatter(K[-1],A[-1],alpha=1,c='red',label='end',marker='x') #plotting points at the start and end of the walk
  plt.legend() #our legend
  plt.grid() #grid lines
  plt.show() #to show

proj(1000) #calling our function with 1000 steps
```

Now that we've got some brute force functions, were going to change focus to using numpy to create scatterplots comparing the end values. This distance form the origin should on average yield the $\sqrt{N}$ result, $N$ being the number of steps taken.

Below we will plot these endpoints:


```python
import numpy as np
import matplotlib.pyplot as plt

#first lets set up the parameters
num_steps = 1000
num_walks = 1000
step_size = 1

#then initialize the position matrix, having rows the size of 2 and collums the size of the number of walks we take. This pos. matrix should hold the start and endpoints of the random walk
pos = np.zeros((num_walks, 2))

#next preform the random walk
for i in range(num_steps):
    #generate a random vector (of length 1) for each step
    direction = np.random.rand(num_walks, 2) - 0.5
    norm = np.linalg.norm(direction, axis=1)
    direction = direction / np.expand_dims(norm, axis=1)

    #add the step to the current position with an option to change step size later
    pos += step_size * direction

#now we need to find the distances from the origin and then calculate the average
distance = np.linalg.norm(pos, axis=1)
average=np.mean(distance)

print("Here we expect to see an average distance of {} from the origin after {} steps but instead we see {}.".format(np.sqrt(num_steps),num_steps,average))

# Plot the resulting positions
plt.scatter(pos[:, 0], pos[:, 1])
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()

```

```python

import numpy as np
import matplotlib.pyplot as plt

#set up the parameters
num_steps = 1000
num_walks = 10

#initialize the position matrix
pos = np.zeros((num_walks, num_steps+1))

#perform the random walk
for i in range(num_steps):
    #generate a random direction (-1 or 1) for each step
    direction = np.random.choice([-1, 1], size=num_walks)

    #update the position matrix
    pos[:, i+1] = pos[:, i] + direction

    #apply a wall constraint
    pos[pos[:, i+1] < 0, i+1] = 0

#plot the resulting positions
plt.xlabel('Steps')
plt.ylabel('Distance from 0')
plt.grid()
plt.plot(pos.T)
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt

#set up the parameters
num_steps = 1000
num_walks = 100
step_size = 1
gravity = 0.1

#initialize the position matrix
pos = np.zeros((num_walks, 2))

#perform the random walk
for i in range(num_steps):
    #generate a random vector (of length 1) for each step
    direction = np.random.rand(num_walks, 2) - 0.5
    norm = np.linalg.norm(direction, axis=1)
    direction = direction / np.expand_dims(norm, axis=1)

    #apply the gravity and step size
    direction[:, 1] -= gravity
    direction *= step_size

    #update the position matrix
    pos += direction

    #apply the floor constraint
    pos[pos[:, 1] < 0, 1] = 0
    pos[:, 1] = np.maximum(pos[:, 1], 0)

#plot the resulting positions
plt.scatter(pos[:, 0], pos[:, 1])
plt.xlabel('Floor')
plt.ylabel('Height')
plt.show()
```
